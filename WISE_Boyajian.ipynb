# Near-IR photometry for Boyajian's Star from WISE

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
### Reading in data 

mjd, w1mpro, w1mpro_err, w2mpro, w2mpro_err = np.loadtxt('boyajian_table.txt', usecols=(7,8,9,11,12), unpack=True)
```

```python
### Sorting columns into ascending order

mjd_all = np.sort(mjd)
w1 = np.sort(w1mpro)
w2 = np.sort(w2mpro)
```

Full light curves: 

```python
plt.plot(mjd_all,w1,'b-')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')

### Blue curve = w1mpro band
```

```python
plt.plot(mjd_all,w2,'g-')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')

### Green curve = w2mpro band
```

Sub light curves for each visit. 1st visit = A, 2nd visit = B

```python
curve_a = np.where((mjd > 55300) & (mjd < 55350))
curve_b = np.where((mjd > 55500) & (mjd < 55550))
```

```python
mjd_a = np.sort(mjd[curve_a])
w1_a = np.sort(w1mpro[curve_a])
```

```python
plt.plot(mjd_b,w1_b,'b-')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')
plt.title('Visit B, w1mpro band')
```
