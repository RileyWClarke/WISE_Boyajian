# Near-IR photometry for Boyajian's Star from WISE

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
```

```python
### Reading in data 

mjd, w1mpro, w1mpro_err, w2mpro, w2mpro_err = np.loadtxt('boyajian_table.txt', usecols=(7,8,9,11,12), unpack=True)
w1_neo, w1_err_neo, w2_neo, w2_err_neo, neo_mjd = np.loadtxt('neoWISE.txt', usecols=(5,6,9,10,18), unpack=True)
```

## ALLWISE Data

Full light curves: 

```python
plt.plot(mjd,w1mpro,'b.')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')

### Blue curve = w1mpro band
```

```python
plt.plot(mjd,w2mpro,'g.')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')

### Green curve = w2mpro band
```

Sub light curves for each visit. 1st visit = A, 2nd visit = B

```python
visit_a = np.where((mjd > 55300) & (mjd < 55350))
visit_b = np.where((mjd > 55500) & (mjd < 55550))
```

```python
mjd_a = mjd[visit_a]
mjd_b = mjd[visit_b]
w1_a = w1mpro[visit_a]
w1_b = w1mpro[visit_b]
w2_a = w2mpro[visit_a]
w2_b = w2mpro[visit_b]
```

```python
plt.plot(mjd_a,w1_a,'b.')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')
plt.title('Visit A, w1mpro band')
```

```python
plt.plot(mjd_b,w1_b,'b.')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')
plt.title('Visit B, w1mpro band')
```

```python
plt.plot(mjd_a,w2_a,'g.')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')
plt.title('Visit A, w2mpro band')
```

```python
plt.plot(mjd_b,w2_b,'g.')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')
plt.title('Visit B, w2mpro band')
```

## NEOWISE data

```python
###w1mpro band, all visits
plt.plot(neo_mjd,w1_neo,'b.')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')

### From left to right, visits C, D, E, F, G, H
```

```python
###w2mpro band, all visits
plt.plot(neo_mjd,w2_neo,'g.')
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')
```

```python
visit_c = np.where((neo_mjd > 56700) & (neo_mjd < 56800))
visit_d = np.where((neo_mjd > 56900) & (neo_mjd < 57000))
visit_e = np.where((neo_mjd > 57100) & (neo_mjd < 57200))
visit_f = np.where((neo_mjd > 57300) & (neo_mjd < 57400))
visit_g = np.where((neo_mjd > 57500) & (neo_mjd < 57600))
visit_h = np.where((neo_mjd > 57600) & (neo_mjd < 57800))
```

```python
mjd_c = neo_mjd[visit_c]
mjd_d = neo_mjd[visit_d]
mjd_e = neo_mjd[visit_e]
mjd_f = neo_mjd[visit_f]
mjd_g = neo_mjd[visit_g]
mjd_h = neo_mjd[visit_h]
w1_c = w1_neo[visit_c]
w1_d = w1_neo[visit_d]
w1_e = w1_neo[visit_e]
w1_f = w1_neo[visit_f]
w1_g = w1_neo[visit_g]
w1_h = w1_neo[visit_h]
w2_c = w2_neo[visit_c]
w2_d = w2_neo[visit_d]
w2_e = w2_neo[visit_e]
w2_f = w2_neo[visit_f]
w2_g = w2_neo[visit_g]
w2_h = w2_neo[visit_h]
```

```python
fig = plt.figure(figsize = (15,10))
fig.suptitle("NEOWISE W1MPRO band", fontsize=14)

fig.add_subplot(231)   
  
plt.plot(mjd_c,w1_c,'b.')
plt.ylim(10.30,10.52)
plt.xlabel('mjd (days)')
plt.ylabel('Single-exposure magnitude')
plt.title('Visit C')

fig.add_subplot(232) 

plt.plot(mjd_d,w1_d,'b.')
plt.ylim(10.30,10.52)
plt.xlabel('mjd (days)')

plt.title('Visit D')

fig.add_subplot(233) 

plt.plot(mjd_e,w1_e,'b.')
plt.ylim(10.30,10.52)
plt.xlabel('mjd (days)')

plt.title('Visit E')

fig.add_subplot(234)

plt.plot(mjd_f,w1_f,'b.')
plt.ylim(10.30,10.52)
plt.title('Visit F')
plt.ylabel("Single Exposure Magnitude")

fig.add_subplot(235)

plt.plot(mjd_g,w1_g,'b.')
plt.ylim(10.30,10.52)
plt.title('Visit G')

fig.add_subplot(236)

plt.plot(mjd_h,w1_h,'b.')
plt.ylim(10.30,10.52)
plt.title('Visit H')

```
